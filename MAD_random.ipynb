```python
import numpy as np
import matplotlib.pyplot as plt
import random

# on initialise un seuil
threshold = 2.5

# On initialise une liste de 30 points entre 200 et 230
length_points = 50
points = np.random.uniform(200, 230, length_points)

plt.plot(points, "-gD")
plt.show()
```

```python
# On génère des outliers
indices_outliers = list(np.random.randint(0, length_points, 4))
outliers = list(np.random.randint(50, 200, 20)) + list(np.random.randint(230, 350, 20))

# on insère des outliers aux indices générés
for indice in indices_outliers:
    points[indice] = random.choice(outliers)
    
print("indices des outliers: " + str(indices_outliers))
print("valeurs associées: " + str([int(points[i]) for i in indices_outliers]))
    
plt.plot(points, "-gD", markevery=indices_outliers, markerfacecolor='red', markeredgecolor = "red")
plt.show()
```

```python
# Methode de l'écart moyen absolu
def median_absolute_deviation(points):   
    median = np.median(points)
    median_removed = np.abs(points - median)
    median_deviation = np.median(median_removed)
    return list(median_removed / median_deviation)

statistic = median_absolute_deviation(points)

# on récupère les indices de points où les "outliers" dépassent le seuil fixé
outliers = []
for i in range(len(statistic)):
    if statistic[i] > threshold:
        outliers.append(i)

plt.plot(statistic, '-gD', markevery=outliers, markerfacecolor='red', markeredgecolor = "red")
plt.show()
```
